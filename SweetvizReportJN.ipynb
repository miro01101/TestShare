```python
import sweetviz
import pandas as pd
```

```python
df = pd.read_csv('http://cooltiming.com/SV/BankChurners_clean.csv')
df.head()
```

```python
report_1 = sweetviz.analyze(df, "Attrition_Flag")
report_1.show_html()
```

```python
report_2 = sweetviz.compare_intra(df, df["Total_Ct_Chng_Q4_Q1"] < 0.6, ["Low_Ct_Chng_Q1Q4", "High_Ct_Chng_Q1Q4"], "Attrition_Flag")
report_2.show_html()
```
